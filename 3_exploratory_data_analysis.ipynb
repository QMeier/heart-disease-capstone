```python
from scipy.stats import chi2_contingency

# Define the categorical columns including 'Num_Major_Vessels' as categorical
categorical_features = [
    'Sex', 'Chest_Pain_Type', 'Fasting_BS', 'Resting_ECG',
    'Exercise_Induced_Angina', 'ST_Slope', 'Thal_Test', 'Disease', 'Num_Major_Vessels'
]

df_chi2 = pd.DataFrame(df, columns = categorical_features)


chi_squared_results = []

# Perform chi-squared tests on each pair of categorical variables
for i in range(len(categorical_features)):
    for j in range(i + 1, len(categorical_features)):
        contingency_table = pd.crosstab(df_chi2[categorical_features[i]], df_chi2[categorical_features[j]])
        chi2, p_value, dof, _ = chi2_contingency(contingency_table, correction=False)
        chi_squared_results.append({
            'Variable 1': categorical_features[i],
            'Variable 2': categorical_features[j],
            'Chi-Square Statistic': chi2,
            'P-Value': p_value,
            'Degrees of Freedom': dof
        })

# Convert the list of results into a DataFrame
chi_squared_df = pd.DataFrame(chi_squared_results)

# Optionally, sort by P-Value to see the most significant results first
chi_squared_df.sort_values(by='P-Value', inplace=True)

# Display the results
print(chi_squared_df)
chi_squared_df.sort_values(by='P-Value')  # Sort by P-Value to see the most significant results first
```
